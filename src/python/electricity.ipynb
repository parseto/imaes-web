```python
import numpy as np
import pandas as pd

from matplotlib import rc, font_manager, rcParams
from matplotlib import pyplot as plt

font_name = font_manager.FontProperties(fname = "c:/Windows/fonts/malgun.ttf").get_name()
rc('font', family = font_name)
rcParams['axes.unicode_minus'] = False

import seaborn as sns

import plotly.graph_objects as go
```

```python
# random data generate
from datetime import datetime
from dateutil.relativedelta import relativedelta

dates = pd.date_range(start = datetime(2020, 1, 1), end = datetime(2024, 8, 1), freq = 'MS')
values = np.random.randint(low = 100, high = 1000, size = len(dates))

data = pd.DataFrame({"date" : dates
                    ,"values" : values})
```

```python
def test_plot():
    plt.figure(figsize = (6, 4))
    plt.plot(data["date"], data["values"])
    plt.title("Test")
    plt.savefig("../images/forecast/electricity-forecast.png", bbox_inches = "tight")
```

```python
import plotly.graph_objects as go

def plotly_test_plot(data):
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=data["date"],
        y=data["values"],
        mode='lines+markers',
        name='Values'
    ))

    fig.update_layout(
        title="Test",
        xaxis_title="Date",
        yaxis_title="Values",
        template="plotly_white"
    )

    # HTML 파일로 저장
    fig.write_html("../images/forecast/electricity-forecast.html")
```

```python
plotly_test_plot(data)
```

```python
test_plot()
```
